# Plotting a bar graph of male vs female buyers participated in the sales.

```python
gc = ds['Buyer Gender'].value_counts()               # to find the number males and females participated in the sales
```

```python
gc
```

```python
sns.set(font_scale=1.4)
ds['Buyer Gender'].value_counts().plot(kind='bar', figsize=(8, 8), rot=0, color='green')
plt.xlabel("Gender", labelpad=14)
plt.ylabel("Count of People", labelpad=14)
plt.title("Bar graph of male vs female buyers participated in the sales", y=1.02);
```
